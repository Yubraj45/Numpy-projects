```python
train_clean= train.copy()
test_clean= test.copy()
train_clean['Age'] = train_clean.groupby(['Pclass', 'Sex'])['Age'].transform(lambda x: x.fillna(x.median()))
test_clean['Age']= test_clean.groupby(['Pclass', 'Sex'])['Age'].transform(lambda x: x.fillna(x.median()))
```

```python
train_clean
```

```python
train_clean.isnull().sum()
```

```python
train_clean['Embarked']= train_clean['Embarked'].fillna(train_clean['Embarked'].mode()[0])
```

```python
test_clean['Fare']= test_clean['Fare'].fillna(test_clean['Fare'].median())
```

```python
test_clean.isnull().sum()
```

```python
train_clean['Has_Cabin']= train_clean['Cabin'].notnull().astype(int)
test_clean['Has_Cabin']= test_clean['Cabin'].notnull().astype(int)
```

```python
train_clean['Deck'] = train_clean['Cabin'].str[0]
test_clean['Deck'] = test_clean['Cabin'].str[0]
```

```python
train_clean['Deck'] = train_clean['Deck'].fillna('U')
test_clean['Deck'] = test_clean['Deck'].fillna('U')
```

```python
train_clean.drop('Cabin', axis=1, inplace=True)
test_clean.drop('Cabin', axis=1, inplace=True)
```

```python
train_clean['Title']= train_clean['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
test_clean['Title']= test_clean['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
rare_titles = ['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 'Major',
               'Rev', 'Sir', 'Jonkheer', 'Dona']
```

```python
train_clean['Title']= train_clean['Title'].replace(rare_titles, 'Rare')
test_clean['Title']= test_clean['Title'].replace(rare_titles, 'Rare')
```

```python
#CLEAN SOME TITLES
train_clean['Title'] = train_clean['Title'].replace(['Mlle', 'Ms'], 'Miss')
train_clean['Title'] = train_clean['Title'].replace('Mme', 'Mrs')

test_clean['Title'] = test_clean['Title'].replace(['Mlle', 'Ms'], 'Miss')
test_clean['Title'] = test_clean['Title'].replace('Mme', 'Mrs')
```

```python
#FAMILY SIZE
train_clean['FamilySize']= train_clean['Parch'] + train_clean['SibSp'] + 1
test_clean['FamilySize']= test_clean['Parch'] + test_clean['SibSp'] + 1
```

```python
train_clean['IsAlone']= (train_clean['FamilySize']==1).astype(int)
test_clean['IsAlone']= (test_clean['FamilySize']==1).astype(int)
```

```python
train_clean.isnull().sum()
```

```python
test_clean.isnull().sum()
```

```python
train_clean['Sex']= train_clean['Sex'].map({'male':0, 'female':1})
test_clean['Sex']= test_clean['Sex'].map({'male':0, 'female':1})

train_clean= pd.get_dummies(train_clean, columns=['Embarked', 'Title', 'Deck'], drop_first=True)
test_clean= pd.get_dummies(test_clean, columns=['Embarked', 'Title', 'Deck'], drop_first=True)
```

```python
train_clean[train_clean.select_dtypes(include='bool').columns]= train_clean.select_dtypes(include='bool').astype(int)
test_clean[test_clean.select_dtypes(include='bool').columns]= test_clean.select_dtypes(include='bool').astype(int)
```
